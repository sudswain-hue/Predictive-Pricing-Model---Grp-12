```python
# Removes finds all sample and saves to new df
LostTrsct = LostTrsct[LostTrsct['Item'] == 'Sample']
```

```python
print(LostTrsct.shape)
LostTrsct.head()
```

```python
# Subset of the DataFrame containing only numeric columns
num_df = LostTrsct.select_dtypes(include='number')

# Identify which numeric columns have null values
cols_with_nulls = num_df.columns[num_df.isnull().any()]

# Replace nulls in each numeric column with that column's median
for col in cols_with_nulls:
    median_val = LostTrsct[col].median()
    LostTrsct[col].fillna(median_val, inplace=True)

# (Optional) Verify the changes
print(LostTrsct.info())
print(LostTrsct.isnull().sum())
```

```python
# Dropping columns Item and Description
LostTrsct = LostTrsct.drop(columns=['Item','Description (Items)','Record Id (Items)', 'Record Id', 'Deal Name', 'Amount'])
LostTrsct.shape
```

```python
LostTrsct = LostTrsct.dropna()
LostTrsct.shape
```

```python
LostTrsct['Label'] = 0
print(LostTrsct.shape)
LostTrsct.head()
```

```python
# Parse consumer B2B etc

def clean_respondent_type(val):
    # Handle missing or blank values.
    if pd.isna(val) or val.strip() == "" or val.strip().lower() == "blank":
        return None

    # Normalize text by stripping whitespace and converting to lower-case.
    val_std = val.strip().lower()

    # Create flags for keywords.
    has_b2b = "b2b" in val_std
    has_consumer = "consumer" in val_std
    has_healthcare = "healthcare" in val_std
    has_technology = "technology" in val_std

    # Apply rules:
    # Rule 4: Both B2B and Consumer --> "Consumer/B2B"
    if has_b2b and has_consumer:
        return "Consumer/B2B"
    # Rule 1: B2B without Consumer --> "B2B"
    elif has_b2b and not has_consumer:
        return "B2B"
    # Rule 2: Consumer without B2B --> "Consumer"
    elif has_consumer and not has_b2b:
        return "Consumer"
    # Rule 3: Healthcare only --> "Healthcare"
    elif has_healthcare and (not has_b2b) and (not has_consumer):
        return "Healthcare"
    # Rule 4: Technology only --> "Healthcare"
    elif has_technology and (not has_b2b) and (not has_consumer):
        return "Technology"
    else:
        # If nothing matches, return the original value.
        return val

# Apply the cleaning function to the "Respondent Type" column and store results in a new column
LostTrsct["Respondent_Cleaned"] = LostTrsct["Respondent Type"].apply(clean_respondent_type)

# Permanently update LostTrsct by keeping only rows where Cleaned_Category is not None.
LostTrsct = LostTrsct[LostTrsct["Respondent_Cleaned"].notnull()].copy()

# Display the updated DataFrame
LostTrsct.sample(10)
```

```python
# Filtering out rows that have technology
LostTrsct = LostTrsct.drop('Respondent Type', axis=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure your DataFrame is loaded into combined_df
# For example: combined_df = pd.read_csv("your_file.csv")
# Check the DataFrame columns
print("Columns in combined_df:", combined_df.columns.tolist())

# Define the numerical columns you expect.
numeric_cols = ['Qty', 'IR', 'LOI', 'Customer Rate', 'Item Amount']

# Convert the columns in numeric_cols to numeric values, coercing errors to NaN.
for col in numeric_cols:
    combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')

# Optionally, print out the data types to verify the conversion:
print(combined_df[numeric_cols].dtypes)

# Create a subplot for each numeric variable in one row.
n_cols = len(numeric_cols)
fig, axes = plt.subplots(nrows=1, ncols=n_cols, figsize=(4 * n_cols, 4))

# If there's only one column, ensure axes is a list.
if n_cols == 1:
    axes = [axes]

# Generate a boxplot for each numeric column.
for ax, col in zip(axes, numeric_cols):
    # Drop NaN values for a clean plot.
    data = combined_df[col].dropna()
    if data.empty:
        print(f"No valid numeric data for column {col}")
    else:
        ax.boxplot(data)
    ax.set_title(f'Boxplot of {col}')
    ax.set_ylabel(col)

plt.tight_layout()
plt.show()
```

```python


# Here, new_df already exists and has the columns:
# "Account Name", "Client Segment Type", "Respondent_Cleaned", "Account Owner",
# "Qty", "IR", "LOI", "Customer Rate", "Item Amount", "Label"

# Verify the current columns if needed:
print("Columns in new_df:", new_df.columns.tolist())

# Define the numerical columns that you want to use for outlier removal.
numeric_cols = ['Qty', 'IR', 'LOI', 'Customer Rate', 'Item Amount']

# Create a working copy of new_df so the original new_df remains unchanged if needed.
df_clean = new_df.copy()

# Convert the numeric columns to actual numeric types; non-numeric values become NaN.
for col in numeric_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# (Optional) Check conversion by printing data types for the numeric columns:
print("Data types after conversion:")
print(df_clean[numeric_cols].dtypes)

# Loop over each numeric column and filter out the outliers (using the IQR method).
for col in numeric_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Keep rows where the column's value is within the calculated bounds.
    df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

# Remove rows where 'Qty', 'Customer Rate', or 'Item Amount' are equal to 0.
df_clean = df_clean[(df_clean['Qty'] != 0) &
                    (df_clean['Customer Rate'] != 0) &
                    (df_clean['Item Amount'] != 0)]

# Reorder the columns to the desired order.
new_order2 = [
    "Account Name",
    "Client Segment Type",
    "Respondent_Cleaned",
    "Account Owner",
    "Qty",
    "IR",
    "LOI",
    "Customer Rate",
    "Item Amount",
    "Label"
]
df_clean = df_clean[new_order2]

# Optionally, display a sample of the cleaned DataFrame.
df_clean.sample(10)
```
